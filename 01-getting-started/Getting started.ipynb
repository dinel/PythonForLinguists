# Why Python?

- Surveys list Python as one of the top programming languages (for example a 2021 [IEEE survey](https://www.theregister.com/2021/08/25/top_programming_language_python/)) and very good to have in your CVs
- It is interpreted (quick in terms of development, but can be slow at running time)
- It is easy to use (5 times shorter than equivalent in Java and 10 times shorter than equivalent in C)
- Easy to learn, easy to read, clean structure: Python 3 has 33 keywords, C++ has 62, Java has 53, and Visual Basic has more than 120 (the actual number varies depending on the implementation)
- It is powerful
- It runs everywhere
- It has a strong community
- It is free and open source

(it is object oriented, it is a "glue" language, it is fashionable, ...)


# Why we use Python in Natural Language Processing/Translation Technologies

- There is the [NLTK library](http://nltk.org) (and now there is [spaCy](https://spacy.io/))
- Very good libraries for processing text and for regular expressions
- Python is widely used in data science (**very good for employability!!**)
- There are numerous libraries which are useful: ``NumPy`` and ``SciPy`` for numerical calculations, ``Requests`` for accessing APIs and retrieving data from Internet, ``BeautifulSoup`` for HTML parsing, ``RegEx`` for regular expressions
- There are various libraries for machine learning and deep learning: scikit-learn, TensorFlow, Keras, ... 


# History of Python

- Developed in the late 1980s by Guide Van Rossum and currently maintained by a core development team
- Version 1.0 was released in 1994
- Version 2.0 was released in 2000
- Version 3.0 was released in 2008 and was not backwards compatible with 2.x
- Python 2.7 is no longer maintained

You can read about the history of Python on Wikipedia: https://en.wikipedia.org/wiki/History_of_Python

The name of the programming language is not derived from the snake, but from Monty Python's Flying Circus :)

# Using python

### Interactive mode (REPL = Read-eval-print loop)
   - the quickest way to interact with python. 
   - tell python what to do and it responds immediately
   - it can be started by calling ``python`` directly or by relying on your IDE
   - it is usually indicated by the ``>>>`` prompt or in Spyder using ``In [1]`` in the IPython console.
   
### Script mode
   - you can write, edit, load, save your code
   - you run your code only when you feel it is ready
   - any text editor can be used to edit your code
   - to run the code, the name of the file has to be passed as an argument to python command
   - however, using an IDE makes the process much simpler (and usually it has syntax highlighting)
